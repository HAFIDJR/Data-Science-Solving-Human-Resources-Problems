```python
def categorical_plot(features,df,segment_feature = None ):
  fig , ax = plt.subplots(len(features) ,1,figsize=(8,15))
  for i , feature in enumerate(features):
    if segment_feature:
      sns.countplot(data = df , x = segment_feature , hue = feature , ax = ax[i])
    else:
      sns.countplot(data = df , x = feature , ax = ax[i])

  plt.tight_layout()
  plt.show()
```

```python
categorical_plot(
    features = ['BusinessTravel', 'Department', 'EducationField', 'Gender',
       'MaritalStatus', 'Over18', 'OverTime'],
    df = df,
    segment_feature = 'status_employe'
)
```

```python
def numerical_dis_plot(features,df,segment_feature = None ,showfliers=True):
  fig, ax = plt.subplots(len(features), 1,figsize=(8,15))
  for i, feature in enumerate(features):
        if segment_feature:
            sns.boxplot(y=segment_feature, x=feature, data=df ,ax=ax[i], showfliers=showfliers)
            ax[i].set_ylabel(None)
        else:
            sns.boxplot(x=feature, data=df, ax=ax[i], showfliers=showfliers)
  plt.tight_layout()
  plt.show()
```

```python
# menampilkan Korelasi antara numerical value dangan attraction
numerical_dis_plot(
    features = ['Age' ,'HourlyRate' ,'MonthlyIncome','PercentSalaryHike','TotalWorkingYears','YearsAtCompany','DailyRate','DistanceFromHome','RelationshipSatisfaction'],
    df = df,
    segment_feature = 'status_employe'
)
```

```python
#mengubah value menjadi interval
def conv_to_interval(label):
    return f'{label.left} - {label.right}'

#mengubah data numerik menjadi kategorik
def conv_to_categorical(df,column,num):
    bins = np.linspace(0,df[column].max(),num=num).round()
    return pd.cut(df[column],bins = bins).apply(lambda x : conv_to_interval(x))
```

```python
df['MonthlyIncome'] = conv_to_categorical(df,'MonthlyIncome',5)
df['OverTime'] = df['OverTime'].apply(lambda x : x == 'Yes')
```

```python
# Mengidentifikasi pendapatan bulanan terendah
lowest_income = df['MonthlyIncome'].unique()[0]

# Mengelompokkan data berdasarkan JobRole dan menghitung metrik agregasi
job_agg_stats = df.groupby('JobRole').agg(
    median_job_level=('JobLevel', 'median'),
    median_years_at_company=('YearsAtCompany', 'median'),
    low_income_percentage=('MonthlyIncome', lambda x: (x.value_counts()[lowest_income] / len(x)) * 100),
    poor_relationship_percentage=('RelationshipSatisfaction', lambda x: (x.value_counts()[1] / len(x)) * 100),
    low_job_satisfaction_percentage=('JobSatisfaction', lambda x: (x.value_counts()[1] / len(x)) * 100),
    poor_work_life_balance_percentage=('WorkLifeBalance', lambda x: (x.value_counts()[1] / len(x)) * 100),
    unsatisfactory_environment_percentage=('EnvironmentSatisfaction', lambda x: (x.value_counts()[1] / len(x)) * 100),
    attrition_rate=('Attrition', lambda x: x.mean() * 100),
    overtime_rate=('OverTime', lambda x: x.mean() * 100),
)

# Mengganti nama kolom untuk meningkatkan keterbacaan
job_agg_stats.rename(
    columns={
        'median_job_level': 'Median Job Level',
        'median_years_at_company': 'Median Years at Company',
        'low_income_percentage': 'Low Income (%)',
        'poor_relationship_percentage': 'Poor Relationship (%)',
        'low_job_satisfaction_percentage': 'Low Job Satisfaction (%)',
        'poor_work_life_balance_percentage': 'Poor Work-Life Balance (%)',
        'unsatisfactory_environment_percentage': 'Unsatisfactory Environment (%)',
        'attrition_rate': 'Attrition Rate (%)',
        'overtime_rate': 'Overtime Rate (%)',
    },
    inplace=True
)
```

## Data Preparation / Preprocessing

```python
X = df.drop(columns=['Attrition','status_employe'])
y = df['Attrition']
```

```python
y
```

```python
# memilih X dengan menghapus data dengan type selain numerik
X = X.select_dtypes(include='number')
```

```python
X.info()
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
cols = X_train.columns
```

```python
# preprocessing
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])
```

## Modeling

```python
# import KNeighbors ClaSSifier from sklearn
from sklearn.neighbors import KNeighborsClassifier


# instantiate the model
knn = KNeighborsClassifier(n_neighbors=3)


# fit the model to the training set
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(knn.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(knn.score(X_test, y_test)))
```

```python
# Mencoba mengubah nilai n_neigboar
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
# Plot error_rate
plt.figure(figsize=(10,6))
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt
```

```python
# instantiate the model
knn_5 = KNeighborsClassifier(n_neighbors=3)


# fit the model to the training set
knn_5.fit(X_train, y_train)
```

```python
y_pred = knn_5.predict(X_test)

y_pred
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(knn_5.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(knn_5.score(X_test, y_test)))
```

## Evaluation

```python
# Confusion Matriks
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```
